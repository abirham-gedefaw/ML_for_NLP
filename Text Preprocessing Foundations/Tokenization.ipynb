```python
import nltk
```

```python
corpus = """In 2026, climate scientists are increasingly turning to Artificial Intelligence (AI) to optimize global power grids.
Dr. Elena Rostova, a leading researcher at the Euro-Mediterranean Center on Climate Change, stated that machine learning algorithms have successfully reduced carbon emissions by 14% in pilot cities.
However, implementing these systems worldwide requires massive computational power, sparking a fierce debate over the environmental footprint of data centers.
Can technology truly solve the very crisis it helps accelerate?"""
```

```python
print(corpus)
```

```python
# Tokenization
# paragraph ---> sentences

from nltk.tokenize import sent_tokenize

try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')

documents = sent_tokenize(corpus)

documents
```

```python
type(documents)
```

```python
for sentences in documents:
  print(sentences)
```

```python
# Tokenization
# paragraph ---> words
# sentence ---> words
from nltk.tokenize import word_tokenize
words = word_tokenize(corpus)
words
```

```python
# Other libraray with a litle bit difference.
from nltk.tokenize import wordpunct_tokenize
words = word_tokenize(corpus)
words
```

```python
# Other libraray with a litle bit difference.
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
words = tokenizer.tokenize(corpus)
words
```
