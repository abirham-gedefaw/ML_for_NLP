```python
paragraph =( """The primary goal of building an effective Natural Language Processing pipeline is to cleanly extract the most meaningful words out of a massive wall of text.
When we are training a machine learning model, 
common words like 'the', 'is', 'at', and 'which' appear so frequently that they can easily overwhelm our algorithms.
These filler words are officially called stopwords, and they usually do not add any 
valuable context to the underlying sentiment of a sentence.
Therefore, we almost always filter them completely out during our initial preprocessing phase.
By stripping away these redundant elements, we can significantly reduce 
the overall size of our dataset, which directly allows our neural networks 
to focus entirely on the core keywords that actually matter.
"""
)
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
stop_words
```

```python
sentences = sent_tokenize(paragraph)
sentences
```

```python
## Apply stopwords and filter then apply stemming

stemmer = PorterStemmer()

for i in range(len(sentences)):
  words = word_tokenize(sentences[i])
  cleaned_words = [stemmer.stem(word) for word in words if word.lower() not in stop_words]
  sentences[i] = ' '.join(cleaned_words)
sentences
```

```python
## Apply stopwords and filter then apply snowball stemming

snowballstemmer = SnowballStemmer("english")

for i in range(len(sentences)):
  words = word_tokenize(sentences[i])
  cleaned_words = [snowballstemmer.stem(word) for word in words if word.lower() not in stop_words]
  sentences[i] = ' '.join(cleaned_words)
sentences
```

```python
## Apply stopwords and filter then apply lemmatization

lemmatizer = WordNetLemmatizer()

for i in range(len(sentences)):
  words = word_tokenize(sentences[i])
  cleaned_words = [lemmatizer.lemmatize(word) for word in words if word.lower() not in stop_words]
  sentences[i] = ' '.join(cleaned_words)
sentences
```
